```python
import pandas as pd
%matplotlib inline
```

```python
df = pd.read_csv('Data/starbucks.csv', index_col='Date', parse_dates=True)
```

```python
df.head()
```

```python
df['Close'].plot(figsize=(12,5))
```

```python
df['Close'].plot(figsize=(12, 5))
df.rolling(window=7).mean()['Close'].plot()
```

```python
df['Close: 1Week mean'] = df['Close'].rolling(window=7).mean()
```

```python
df.head(8)
```

```python
df[['Close','Close: 1Week mean']].plot(figsize=(12,5))
```

```python
df['Close'].expanding().mean().plot(figsize=(12,5));
```

```python
df['Close'].mean()
```

```python
title = 'TITLE'
ylabel = 'Y LABEL'
xlabel = 'X LABEL'
```

```python
ax = df['Close'].plot(figsize=(12,5), title=title)
ax.autoscale(axis='both', tight=True)
ax.set(xlabel=xlabel, ylabel=ylabel)
```

```python
df['Close']['2017-01-01':'2017-12-31'].plot(figsize=(12, 4))
```

```python
df['Close'].plot(figsize=(12, 4), xlim=['2017-01-01','2017-12-31'],\
                 ylim=[45, 65], ls='--', c='r')
```

```python
from matplotlib import dates
```

```python
ax = df['Close'].plot(figsize=(12, 4),\
                      xlim =['2017-01-01','2017-03-01'], ylim=[50, 60])
ax.set(xlabel='')
ax.xaxis.set_major_locator(dates.WeekdayLocator(byweekday=0))
ax.xaxis.set_major_formatter(dates.DateFormatter('%a-%B-%d'))
```

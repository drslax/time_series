```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
df = pd.read_csv('Data/airline_passengers.csv', index_col="Month")
```

```python
df.dropna(inplace=True)
```

```python
df.index = pd.to_datetime(df.index)
```

```python
df['6-month-SMA'] = df['Thousands of Passengers'].rolling(window=).mean()
```

```python
df['12-month-SMA'] = df['Thousands of Passengers'].rolling(window=12).mean()
```

```python
df.plot(figsize=(12,10))
```

```python
df['EWMA-6'] = df['Thousands of Passengers'].ewm(span=6).mean()
```

```python
df['EWMA-12'] = df['Thousands of Passengers'].ewm(span=12).mean()
```

```python
df[['Thousands of Passengers', 'EWMA-6', 'EWMA-12']].plot(figsize=(12,10))
```

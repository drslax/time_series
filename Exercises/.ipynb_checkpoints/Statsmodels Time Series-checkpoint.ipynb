```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
%matplotlib inline

df = pd.read_csv('../Data/EnergyProduction.csv', index_col=0, parse_dates=True)
df.head()
```

```python
df.index.freq = 'MS'
```

```python
df.index
```

```python
df.plot(figsize=(12,8))
```

```python
from statsmodels.tsa.holtwinters import ExponentialSmoothing
```

```python
df['SMA12'] = df['EnergyIndex'].rolling(window=12).mean()
```

```python
span = 12
alpha = 2 / (span + 1)
df['SES12'] = ExponentialSmoothing(df['EnergyIndex'])\
                .fit(smoothing_level=alpha,optimized=False).fittedvalues.shift(-1)
```

```python
df['DES12'] = ExponentialSmoothing(df['EnergyIndex'], trend= 'add').fit().fittedvalues.shift(-1)
```

```python
df['TESmul12'] = ExponentialSmoothing(df['EnergyIndex'], trend='mul', seasonal='mul', seasonal_periods=12).fit().fittedvalues
```

```python
df.plot(figsize=(12,8))
```
